 ## Model Building

I am importing all the Essentials for 'Train and Test Split' and Importing the basic required modules like "Logistic Regression and K Nearest Neighbour and Decision Tree".

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
X=df.drop(columns=['species'])    # input data
Y=df['species']    #output data
# split the data into train and test
x_train,x_test,y_train,y_test = train_test_split(X , Y , test_size = 0.30)
```

#### Logistic Reggression

```python
model = LogisticRegression()
model.fit(x_train, y_train)
print("Logistic Reggression Accuracy : ", model.score(x_test,y_test) * 100)
```

```python
model.fit(x_train.values,y_train.values)    # train the model
```

```python
print("Accuracy : ", model.score(x_test,y_test) * 100)
```

#### K-nearest Neighbour

```python
model = KNeighborsClassifier()
model.fit(x_train.values , y_train.values)
print("K-Nearest Neighbors Accuracy : ", model.score(x_test,y_test) * 100)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
print("Accuracy : ", model.score(x_test,y_test) * 100)
```

#### Decision Tree

```python
model =DecisionTreeClassifier()
model.fit(x_train.values,y_train.values)
print('Decision Tree Accuracy : ',model.score(x_test,y_test)*100)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
print('Accuracy : ',model.score(x_test,y_test)*100)
```

## Save your Model

```python
import pickle
filename = 'saved_model.pkl'
pickle.dump(model,open(filename,'wb'))
```

```python
try:
    with open(filename, 'wb') as file:
        pickle.dump(model, file)
    print('model saved successfully')
except Exception as e:
    print(f'Error saving the model: {e}')
```
